# About the Data/Understanding the Data.
This data is all about greenhouse gas (GHG) emissions per stage of the food value chain from land (land use change or LUC) to retail (end-user purchase/acquisition point).

The stages identified from Land to Retail are commonly used to describe stages of global food systems. Note that variation occurs from one food product to the next and within a single food product, for reasons including the region and ecological profile of the system of production, size and type of production system, and more.

The data is limited to the global average (mean) GHG emissions for each food product and stage.

GHG emissions are measured in kg CO2e per kg of food produced.

kg CO2e = kilograms of carbon dioxide equivalents
Each food product is listed with global average GHG emissions from each stage of the food value chain from Land, Farm, Animal Feed, Processing, Transport, Packaging, up to Retail.

# Data Visualization
Data visualization is the representation of data in a graphical or visual format to help understand patterns, relationships, and trends in the data.
In this data we have plotted the following to visualize our data:


1.Distribution plot.
Shows the distribution of a dataset. It displays the probability density function of the data and provides a smooth estimate of the distribution.In our case we have used two columns to show their distribution density i.e Column land use change and column processing.Note that this column cannot compare two or more variables you can only compare a variable at a time.


2.Joint plot-
It combines two different plots in one figure to show the relationship between two variables. Unlike a dist plot, you can compare two variables, in our case we have compared processing and transport.


3.Pair plot.
A graphical representation of the pairwise relationships between variables in a dataset. It displays scatterplots of all possible pairs of variables in a grid format where each row and column represents a different variable, and the diagonal plots show the distribution of that variable.They identify patterns and relationships between variables. They allow us to see how different variables relate to each other, and can help us detect correlations, clusters, outliers, and other patterns in the data.Using our data, we have drawn a pair plot that shows how each and every variable can be compared to each other.


4.Histogram.
Are used to visualize the shape of a distribution and to identify patterns, trends, and outliers in the data. They allow us to quickly see how the data is spread out and whether it is skewed, symmetric, or multimodal. In our case, it shows how the 11 columns to show data distribution.


5.Scatter plot.
They visualize the relationship between two variables and to identify patterns, trends, and outliers in the data. They allow us to see whether there is a correlation or association between the variables, and to estimate the strength and direction of the relationship.In our case we have used this plot to show the relationship between Transport and Land usage variables.


6.Line plot.
Visualizes trends and patterns in the data over time or across different categories. They allow us to see how the data changes over a period of time or how it varies across different categories. In our case the plot shows Global Average GHG Emissions per kg over Food Products.


7.Bar chart.
Used to compare the values of different categories or groups. They allow us to easily see which categories have higher or lower values, and to identify patterns and trends in the data. In our case, it shows the total emmissions of greenhouse gases vs the food products that are emitting this gases.
